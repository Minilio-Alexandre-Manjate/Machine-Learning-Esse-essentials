```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('train.csv')
```

```python
data.head()
```

```python
X=data.iloc[:, :-1]
Y=data.iloc[:,-1]
```

```python
X=pd.DataFrame(X)
Y=pd.DataFrame(Y)
```

```python
def normilize(X):
    u=X.mean(axis=0)
    std=X.std(axis=0)
    return (X-u)/std
```

```python
normilize(X)
```

```python
from sklearn.feature_selection import SelectKBest
```

```python
best_features=SelectKBest()
```

```python
fit=best_features.fit(X,Y)
fit.scores_
```

```python
print(X.shape)
print(Y.shape)
```

```python
dfscores=pd.DataFrame(fit.scores_)
dfcolumns= pd.DataFrame(X.columns)
```

```python
featureScores = pd.concat([dfcolumns, dfscores], axis = 1)
featureScores.columns = ['Features', 'Scores']
```
