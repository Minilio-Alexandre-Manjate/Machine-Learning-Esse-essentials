```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Data Generate x,y
def generateDataset(m):
    X= np.random.randn(m)*10
    noise= np.random.randn(m)
    y= 3*X + 1 + 5*noise
    return(X,y)
```

```python
X,y= generateDataset(100)
print(X.shape, y.shape)
```

```python
def plotData(X,y, color='orange', title='Data'):
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(title)
    plt.scatter(X,y, c=color)
    plt.show()
```

```python
plotData(X,y)
```

```python
print(X.mean(), X.std())
```

```python
#normalize the data
def normalizeData(X):
    X=(X-X.mean())/X.std()
    return X
```

```python
plotData(X,y)
X= normalizeData(X)
plotData(X,y)
```

```python
print(X.std(), X.mean())
```

```python
def trainTestSplit(X, y, split=0.8):
    m = X.shape[0]
    data = np.zeros((m, 2))  
    data[:, 0] = X  
    data[:, 1] = y  
    np.random.shuffle(data)
    
    split_index = int(m * split)
    XT=data[:split_index, 0]
    yT=data[: split_index, 1]
    xt=data[split_index: ,0]
    yt=data[split_index: , 1]
    
    
    return XT, yT, xt, yt
```

```python
XT,yT,xt,yt = trainTestSplit(X, y)
```

```python
print(XT.shape, yT.shape)
print(xt.shape, yt.shape)
```

```python
plt.scatter(XT, yT,color='orange', label='Train data')
plt.scatter(xt, yt, color='blue', label='test data')
plt.title('Train-test Data')
plt.legend()
plt.show
```
