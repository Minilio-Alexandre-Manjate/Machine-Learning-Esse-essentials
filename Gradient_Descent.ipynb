```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x) = (x - 5)^2 + 3
def f(x):
    return (x - 5) ** 2 + 3

x = 0
lr = 0.1
X = np.linspace(-10, 10, 100)
Y = f(X)

plt.plot(X, Y)

for i in range(50):
    grad = 2 * (x - 5)
    y = f(X)  # Calculate y-values for all X
    x = x - lr * grad
    plt.scatter(x, f(x), color='red')  # Plot each step of gradient descent
    print(x)

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent')
plt.show()
```
